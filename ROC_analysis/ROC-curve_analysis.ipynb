# <center>1. –ü–æ—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–¥–∞—á–∏</center>

$^1$ __–ó–æ–Ω–æ–π –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏—è –†–õ–°__ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –æ–±–ª–∞—Å—Ç—å –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞, –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö –∫–æ—Ç–æ—Ä–æ–π –†–õ–° –æ–±–µ—Å–ø–µ—á–∏–≤–∞–µ—Ç –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏–µ –æ–±—ä–µ–∫—Ç–æ–≤ —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—è–º–∏ –ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏—è $P_d$ –∏ –ª–æ–∂–Ω–æ–π —Ç—Ä–µ–≤–æ–≥–∏ $P_{fa}$ –Ω–µ —Ö—É–∂–µ —Ç—Ä–µ–±—É–µ–º—ã—Ö.


–ù–∞–ª–∏—á–∏–µ –æ–±—ä–µ–∫—Ç–∞ –æ–±–æ–∑–Ω–∞—á–∏–º –∫–∞–∫ –≥–∏–ø–æ—Ç–µ–∑–∞ $H_1$, –∞ –µ–≥–æ –æ—Ç—Å—É—Ç—Å—Ç–≤–∏–µ –∫–∞–∫ –≥–∏–ø–æ—Ç–µ–∑–∞ $H_0$. –°–∏–≥–Ω–∞–ª $X$ ‚Äî —ç—Ç–æ —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞. –ü—Ä–µ–¥–ø–æ–ª–æ–∂–∏–º, —á—Ç–æ –º—ã –∏–º–µ–µ–º —É—Å–ª–æ–≤–Ω—ã–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è $(1)$, –∫–æ—Ç–æ—Ä—ã–µ –∏–∑–≤–µ—Å—Ç–Ω—ã.

$$f_{(X|H_0)} (x‚îÇH_0 ), \ f_{(X|H_1)}(x‚îÇH_1)\ \ \ \ (1)$$ 

–û—á–µ–≤–∏–¥–Ω–æ, —á—Ç–æ –º—ã –º–æ–∂–µ–º –≥–æ–≤–æ—Ä–∏—Ç—å –æ —Ç–æ–º, —á—Ç–æ $X$ –∏–º–µ–µ—Ç –æ–¥–∏–Ω–∞–∫–æ–≤–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–ª—è –æ–±–æ–∏—Ö –≥–∏–ø–æ—Ç–µ–∑, –Ω–æ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–µ –æ–∂–∏–¥–∞–Ω–∏—è —É—Å–ª–æ–≤–Ω—ã—Ö —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π $(1)$ –±—É–¥—É—Ç —Ä–∞–∑–ª–∏—á–Ω—ã–º–∏. –¢–∞–∫ –∂–µ —Ä–µ–∞–ª–∏—Å—Ç–∏—á–Ω–æ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–∏—Ç—å, —á—Ç–æ

$$E[X‚îÇH_0]<E[X‚îÇH_1]\ \ \ \ (2)$$

–ö–∞–∫ —É–∂–µ –±—ã–ª–æ —Å–∫–∞–∑–∞–Ω–æ —Ä–∞–Ω–µ–µ, —Å–∏–≥–Ω–∞–ª $X$ —Å—Ä–∞–≤–Ω–∏–≤–∞–µ—Ç—Å—è —Å –Ω–µ–∫–∏–º –ø–æ—Ä–æ–≥–æ–º, –∫–æ—Ç–æ—Ä—ã–π –º—ã –æ–±–æ–∑–Ω–∞—á–∏–ª–∏ –∫–∞–∫ $Œª$. –†–µ—à–µ–Ω–∏–µ –æ —Ç–æ–º –µ—Å—Ç—å –ª–∏ –æ–±—ä–µ–∫—Ç –≤ –ó–û –†–õ–° –∏–ª–∏ –Ω–µ—Ç –æ–±–æ–∑–Ω–∞—á–∏–º –∫–∞–∫ $D_1$ –∏ $D_0$ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ. –î–æ–ø–æ–ª–Ω—è—è –†–∏—Å. 2 $Œª, D_1$ –∏ $D_0$, –ø–æ–ª—É—á–∞–µ–º:

<center>–†–∏—Å. 3 –£—Å–ª–æ–≤–Ω—ã–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–ª–æ—Ç–Ω–æ—Å—Ç–µ–π $(1)$ c –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏–µ–º —É—Ä–æ–≤–Ω—è –ø–æ—Ä–æ–≥–∞ $Œª$ –∏ –¥–∏–∞–ø–∞–∑–æ–Ω–æ–≤ –ø—Ä–∏–Ω—è—Ç–∏—è —Ä–µ—à–µ–Ω–∏–π $D_0$ –∏ $D_1$</center>

–ü–æ–ª—É—á–∞–µ–º —Å–ª–µ–¥—É—é—â–∏–µ —É—Å–ª–æ–≤–Ω—ã–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏:

$$
\begin{array}{l}
P_{d}=P\left(D_{1} \mid H_{1}\right)=\int_{D_{1}} f_{X \mid H_{1}}\left(x \mid H_{1}\right) d x \\
P_{f a}=P\left(D_{1} \mid H_{0}\right)=\int_{D_{1}} f_{X \mid H_{0}}\left(x \mid H_{0}\right) d x
\end{array}   (3)
$$

–£—Å–ª–æ–≤–Ω—ã–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ $(3)$ –º–æ–∂–Ω–æ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å —Ç–∞–∫:
$P_d$ ‚Äì –µ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏—è –æ–±—ä–µ–∫—Ç–∞ –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏, —á—Ç–æ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –≥–∏–ø–æ—Ç–µ–∑–∞ $H_1$, —Ç.–µ. –æ–±—ä–µ–∫—Ç –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ –ó–û.

–í–∏–∑—É–∞–ª–∏–∑–∏—Ä—É—è –∏–Ω—Ç–µ–≥—Ä–∞–ª—ã $(3)$, –ø–æ–ª—É—á–∞–µ–º:

<center>–†–∏—Å. 4 –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π $P_d$ –∏ $P_{fa}$ –∫–∞–∫ –ø–ª–æ—â–∞–¥–∏ –ø–æ–¥ –∫—Ä–∏–≤—ã–º–∏ $(1)$</center>

–û—á–µ–≤–∏–¥–Ω–æ, —á—Ç–æ —Å —É–º–µ–Ω—å—à–µ–Ω–∏–µ–º –ø–æ—Ä–æ–≥–∞ $Œª$ –ø–ª–æ—â–∞–¥—å –ø–æ–¥ –∫—Ä–∏–≤—ã–º–∏ –±—É–¥–µ—Ç –±–æ–ª—å—à–µ, –∞ –∑–Ω–∞—á–∏—Ç, —á—Ç–æ $P_d$ –∏ $P_{fa}$ –±—É–¥—É—Ç —É–≤–µ–ª–∏—á–∏–≤–∞—Ç—å—Å—è.

–ö—Ä–∏–≤–∞—è –ø–æ–∫–∞–∑—ã–≤–∞—é—â–∞—è –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å $P_d$ –∫–∞–∫ —Ñ—É–Ω–∫—Ü–∏—é –æ—Ç $P_{fa}$ –¥–ª—è —Ä–∞–∑–ª–∏—á–Ω—ã—Ö $Œª$ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è $ROC$-–∫—Ä–∏–≤–∞—è (–∞–Ω–≥–ª. *Receiver Operating Characteristic*, —Ä–∞–±–æ—á–∞—è —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∞ –ø—Ä–∏—ë–º–Ω–∏–∫–∞).

–û—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã $X$ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–æ–π –û—Ç–Ω–æ—à–µ–Ω–∏–µ —Å–∏–≥–Ω–∞–ª-—à—É–º (–û–¢–®, –∞–Ω–≥–ª. *Signal-to-Noise Ratio*, —Å–æ–∫—Ä. $SNR$), –∫–æ—Ç–æ—Ä–∞—è –∑–∞–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è –∫–∞–∫:


$$
S N R=10 \log _{10}\left(\frac{P_{\text {signal }}}{P_{\text {noise }}}\right)[dB]  \ \ \ (4)
$$


- $P_{signal}$ ‚Äì –º–æ—â–Ω–æ—Å—Ç—å —Å–∏–≥–Ω–∞–ª–∞;
- $P_{noise}$ ‚Äì –º–æ—â–Ω–æ—Å—Ç—å —à—É–º–∞. 

–ü—Ä–∏–Ω–∏–º–∞—è –≤ $(4)$ $P_{signal}=1$, –∞ $P_{signal}=œÉ^2$ ‚Äì –¥–∏—Å–ø–µ—Ä—Å–∏—è, –ø–æ–ª—É—á–∞–µ–º:

$$
S N R=10 \log _{10}\left(\frac{1}{\sigma^{2}}\right)[d B]\ \ (5)
$$

# <center>2. –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ $ROC$-–∫—Ä–∏–≤—ã—Ö</center>

__–ù–∞—á–∞–ª—å–Ω—ã–µ —É—Å–ª–æ–≤–∏—è:__

$$
\begin{array}{c}
f_{X \mid H_{0}}\left(x \mid H_{0}\right)=\frac{1}{\sqrt{2 \pi \sigma^{2}}} e^{-\frac{x^{2}}{2 \sigma^{2}}} \\
f_{X \mid H_{1}}\left(x \mid H_{1}\right)=\frac{1}{\sqrt{2 \pi \sigma^{2}}} e^{-\frac{(x-1)^{2}}{2 \sigma^{2}}}
\end{array}  \ \ \ (6)
$$

$$
S N R \in[-1,5]\ \ \ (7) \\
\lambda \in[0,1]\ \ \ (8)
$$


–ò–∑ $(6)$ –≤–∏–¥–Ω–æ, —á—Ç–æ –º—ã –¥–µ–ª–∞–µ–º –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏–µ –æ —Ç–æ–º, —á—Ç–æ —É—Å–ª–æ–≤–Ω—ã–µ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –∏–º–µ—é—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ, —Ç.–µ. $X \mid H_{0} \sim \mathcal{N}\left(0, \sigma^{2}\right)$ –∏ $X \mid H_{1} \sim \mathcal{N}\left(1, \sigma^{2}\right)$, –≥–¥–µ $\mathcal{N}\left(\mu, \sigma^{2}\right)$ ‚Äì –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏–µ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–º –æ–∂–∏–¥–∞–Ω–∏–µ–º $\mu$ –∏ –¥–∏—Å–ø–µ—Ä—Å–∏–µ–π $\sigma^{2}$. –†–∞–Ω–µ–µ –º—ã –æ–±–æ–∑–Ω–∞—á–∞–ª–∏ $Œ∑_0, Œ∑_1$ –∫–∞–∫ —É—Å–ª–æ–≤–Ω–æ–µ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã $X$. –ò—Å—Ö–æ–¥—è –∏–∑ $(6)$, –ø–æ–ª—É—á–∞–µ–º, —á—Ç–æ $Œ∑_0= Œï[X‚îÇH_0 ]=0$, –∞ $Œ∑_1= Œï[X‚îÇH_1 ]=1$.

–ò–∑ –¥–∏–∞–ø–∞–∑–æ–Ω–∞ $(7)$ –º—ã –º–æ–∂–µ–º –ø–æ–Ω—è—Ç—å, –∫–∞–∫–æ–π —É –Ω–∞—Å –¥–∏–∞–ø–∞–∑–æ–Ω –¥–∏—Å–ø–µ—Ä—Å–∏–∏ $\sigma^2$. –î–ª—è —ç—Ç–æ–≥–æ –≤—ã—Ä–∞–∑–∏–º –∏–∑ $(5)$ $\sigma$:

$$
\begin{array}{c}
S N R=10 \log _{10}\left(\frac{1}{\sigma^{2}}\right)=10 \log _{10}\left(\sigma^{-2}\right)=-2 \cdot 10 \log _{10} \sigma=-20 \log _{10} \sigma \\
-\frac{S N R}{20}=\log _{10} \sigma \\
\sigma=10^{-\frac{S N R}{20}} \ \ (9)
\end{array}
$$

–ü–æ–¥—Å—Ç–∞–≤–ª—è—è –≥—Ä–∞–Ω–∏—á–Ω—ã–µ —Ç–æ—á–∫–∏ –æ—Ç—Ä–µ–∑–∫–∞ –∏–∑ $(7)$, –ø–æ–ª—É—á–∞–µ–º:

$$
\begin{aligned}
&\sigma \in\left[\left.10^{-\frac{S N R}{20}}\right|_{S N R=5},\left.10^{-\frac{S N R}{20}}\right|_{S N R=-1}\right]\\
&\sigma \in\left[10^{-\frac{5}{20}}, 10^{-\frac{-1}{20}}\right]\\
&\sigma \in\left[\sqrt[4]{\frac{1}{10}}, \sqrt[20]{10}\right]
\end{aligned}
$$



–ó–∞–¥–∞–¥–∏–º –≤–µ–∫—Ç–æ—Ä –∑–Ω–∞—á–µ–Ω–∏–π $SNR$ –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è $ROC$-–∫—Ä–∏–≤–æ–π –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∏–∑ –Ω–∏—Ö:

$$
\overline{{S N R}_{\text {values }}}=(-1,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5)^{T} \text { (11) }
$$


–®–∞–≥ –∏–∑–º–µ–Ω–µ–Ω–∏—è –∑–Ω–∞—á–µ–Ω–∏–π –≤–µ–∫—Ç–æ—Ä–∞ $(11)$ —è –≤–∑—è–ª —Ä–∞–≤–Ω—ã–º $\frac{1}{2}$, —Ç.–µ. $‚àÜ_{SNR}=\frac{1}{2}$.
–û–ø—Ä–µ–¥–µ–ª–∏–º —Ç–µ–ø–µ—Ä—å $ROC$-–∫—Ä–∏–≤—É—é –∫–∞–∫ –ø–∞—Ä–∞–º–µ—Ç—Ä–∏—á–µ—Å–∫—É—é —Ñ—É–Ω–∫—Ü–∏—é $P_d$ –æ—Ç $P_{fa}$, –≥–¥–µ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ –≤—ã—Å—Ç—É–ø–∞–µ—Ç $Œª‚àà[0,1]$. –ò–∑ $(3)$ –∏ $(6)$, –ø–æ–ª—É—á–∞–µ–º:

$$
\text { ROC }_{\text {curve }}(\sigma, \lambda)=\left\{\begin{array}{r}
P_{f a}=\frac{1}{\sqrt{2 \pi \sigma^{2}}} \int_{\lambda}^{+\infty} e^{-\frac{x^{2}}{2 \sigma^{2}}} d x \\
P_{d}=\frac{1}{\sqrt{2 \pi \sigma^{2}}} \int_{\lambda}^{+\infty} e^{-\frac{(x-1)^{2}}{2 \sigma^{2}}} d x
\end{array}\right.
$$

$\sigma$ ‚Äì –∑–∞–¥–∞–µ—Ç –∫–æ–Ω–∫—Ä–µ—Ç–Ω—É—é $ROC$-–∫—Ä–∏–≤—É—é –∏–∑ —Å–µ–º–µ–π—Å—Ç–≤–∞ $ROC$-–∫—Ä–∏–≤—ã—Ö.

–ó–∞–¥–∞–¥–∏–º —Ç–µ–ø–µ—Ä—å –≤–µ–∫—Ç–æ—Ä –∑–Ω–∞—á–µ–Ω–∏–π $Œª$:

$$
\overrightarrow{\lambda_{\text {values }}}=(0,0.01,0.02, \ldots, 0.98,0.99,1)^{T}
$$

–®–∞–≥ –∏–∑–º–µ–Ω–µ–Ω–∏—è –∑–Ω–∞—á–µ–Ω–∏–π –≤–µ–∫—Ç–æ—Ä–∞ $(13)$ —è –≤–∑—è–ª —Ä–∞–≤–Ω—ã–º $\frac{1}{100}$  , —Ç.–µ. $‚àÜ_Œª=\frac{1}{100}$.
–í–µ–∫—Ç–æ—Ä $(13)$ —Å–æ–¥–µ—Ä–∂–∏—Ç $101$ –∑–Ω–∞—á–µ–Ω–∏–µ, —Ç.–µ. –∏–º–µ–µ–º $101$ —Ç–æ—á–µ–∫ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –≥—Ä–∞—Ñ–∏–∫–∏ –±—É–¥—É—Ç –≤–∏–∑—É–∞–ª—å–Ω–æ –≤—ã–≥–ª—è–¥–µ—Ç—å –≥–ª–∞–¥–∫–∏–º–∏.

–ò–º–µ—é—â–∏—Ö—Å—è –¥–∞–Ω–Ω—ã—Ö –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è ROC-–∫—Ä–∏–≤—ã—Ö.


```python
# –ø–æ–¥–∫–ª—é—á–µ–Ω–∏–µ –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫
from scipy.stats import norm
from scipy.misc import derivative
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
%matplotlib inline
```

```python
# –∑–∞–¥–∞–¥–∏–º –≤–µ–∫—Ç–æ—Ä SNR –∏ Œª
SNR_values = np.arange(-1, 5.5, 1/2) 
lambda_values = np.arange(0, 1.01, 1/100) 
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è, –∫–æ—Ç–æ—Ä–∞—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ ùùà –ø–æ –∑–∞–¥–∞–Ω–Ω–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é SNR
def snr_to_sigma(SNR):
    return np.power(10, -SNR/20)
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è, –∫–æ—Ç–æ—Ä–∞—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏—è Pfa –∏ Pd –ø–æ –∑–∞–¥–∞–Ω–Ω—ã–º ùùà –∏ Œª
def roc_curve(sigma, lambda_):
    return 1 - norm.cdf(lambda_/sigma), 1 - norm.cdf((lambda_ - 1)/sigma)
```

```python
# —Å–æ–∑–¥–∞–Ω–∏–µ DataFrame –≤ –∫–æ—Ç–æ—Ä–æ–º –±—É–¥—É—Ç —Ö—Ä–∞–Ω–∏—Ç—å—Å—è –∑–Ω–∞—á–µ–Ω–∏—è Pfa, Pd, SNR
data = []
for SNR in SNR_values:
    for lambda_ in lambda_values:
        Pfa, Pd = roc_curve(snr_to_sigma(SNR), lambda_)
        data.append([Pfa, Pd, str(SNR)])
        
data = pd.DataFrame(data, columns=['Pfa', 'Pd', 'SNR'])
```

```python
fig, ax = plt.subplots(figsize=(18, 14))
sns.set_context('poster')
plt.title('$ROC$-–∫—Ä–∏–≤—ã–µ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –∑–Ω–∞—á–µ–Ω–∏—è $SNR$', fontsize=40)
plt.xlabel('$P_{fa}$', fontsize=40)
plt.xticks(np.arange(0, 0.7, 0.05), fontsize=30)
plt.yticks(np.arange(0.5, 1, 0.05), fontsize=30)
plt.ylabel('$P_d$', fontsize=40, rotation=0, labelpad=40)
sns.lineplot(x='Pfa', y='Pd', hue='SNR', 
             data=data, palette='gist_ncar', ax=ax)

ax.annotate("–í–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏–µ $\lambda$", xy=(0.15, 0.81),  xycoords='data',
            xytext=(90, 150), textcoords='offset points',
            size=30, ha='left', 
            arrowprops=dict(arrowstyle="->", color='black',
                            connectionstyle="arc3,rad=0.1"))
plt.grid(color='black', linestyle='--', linewidth=1, alpha=0.3)

plt.legend(fontsize='small', shadow=True, title='$SNR(dB)$',
           borderpad=0.9, fancybox=True);
```

<center>–†–∏—Å. 5 $ROC$-–∫—Ä–∏–≤—ã–µ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –∑–Ω–∞—á–µ–Ω–∏—è $SNR$.</center>

–ò–∑ –≥—Ä–∞—Ñ–∏–∫–∞ –≤–∏–¥–Ω–æ —Å–ª–µ–¥—É—é—â–µ–µ:

1) –í—Å–µ –∫—Ä–∏–≤—ã–µ –Ω–∞—á–∏–Ω–∞—é—Ç —Å–≤–æ–µ –¥–≤–∏–∂–µ–Ω–∏–µ —Å –ª–∏–Ω–∏–∏ $P_{fa}=0.5$ –∏ –∑–∞–∫–∞–Ω—á–∏–≤–∞—é—Ç –Ω–∞ –ª–∏–Ω–∏–∏ $P_d=0.5$;

2) –ß–µ–º –±–æ–ª—å—à–µ –∑–Ω–∞—á–µ–Ω–∏–µ $SNR$, —Ç–µ–º –±–ª–∏–∂–µ $ROC$-–∫—Ä–∏–≤–∞—è –∫ —Ç–æ—á–∫–µ $(0,1)$;

3) –í—Å–µ –∫—Ä–∏–≤—ã–µ –≤—ã–ø—É–∫–ª—ã –≤–≤–µ—Ä—Ö;

4) $ROC$ - –∫—Ä–∏–≤–∞—è –º–æ–Ω–æ—Ç–æ–Ω–Ω–æ –Ω–µ —É–±—ã–≤–∞–µ—Ç. –ß–µ–º –≤—ã—à–µ –ª–µ–∂–∏—Ç –∫—Ä–∏–≤–∞—è, —Ç–µ–º –ª—É—á—à–µ –∫–∞—á–µ—Å—Ç–≤–æ –†–õ–°.


–ü–µ—Ä–≤–æ–µ –æ–±—ä—è—Å–Ω—è–µ—Ç—Å—è –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–Ω–æ—Å—Ç—å—é –∑–Ω–∞—á–µ–Ω–∏—è –ø–æ—Ä–æ–≥–∞ –¥–µ—Ç–µ–∫—Ç–∏—Ä–æ–≤–∞–Ω–∏—è $(8)$. –ü–æ—Å–º–æ—Ç—Ä–∏–º, –∫–∞–∫ –±—É–¥—É—Ç –≤—ã–≥–ª—è–¥–µ—Ç—å $ROC$-–∫—Ä–∏–≤—ã–µ, –µ—Å–ª–∏ —É–≤–µ–ª–∏—á–∏—Ç—å –¥–∏–∞–ø–∞–∑–æ–Ω $Œª$, –Ω–∞–ø—Ä–∏–º–µ—Ä, –¥–æ $[-10,10]$.

```python
# —Å–æ–∑–¥–∞–Ω–∏–µ DataFrame –≤ –∫–æ—Ç–æ—Ä–æ–º –±—É–¥—É—Ç —Ö—Ä–∞–Ω–∏—Ç—å—Å—è –∑–Ω–∞—á–µ–Ω–∏—è Pfa, Pd, SNR
data2 = []
for SNR in SNR_values:
    for lambda_ in np.arange(-10, 10.1, 1/10) :
        Pfa, Pd = roc_curve(snr_to_sigma(SNR), lambda_)
        data2.append([Pfa, Pd, str(SNR)])
        
data2 = pd.DataFrame(data2, columns=['Pfa', 'Pd', 'SNR'])
```

<center>–†–∏—Å. 6 $ROC$-–∫—Ä–∏–≤—ã–µ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –∑–Ω–∞—á–µ–Ω–∏—è $SNR$ c —É–≤–µ–ª–∏—á–µ–Ω–Ω—ã–º –¥–∏–∞–ø–∞–∑–æ–Ω–æ–º $Œª$.</center>

–ü—Ä–∏ —É–≤–µ–ª–∏—á–µ–Ω–∏–∏ –¥–∏–∞–ø–∞–∑–æ–Ω–∞ –∑–Ω–∞—á–µ–Ω–∏–π, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–∂–µ—Ç –ø—Ä–∏–Ω–∏–º–∞—Ç—å $Œª$ –ø–æ–ª—É—á–∞–µ—Ç—Å—è, —á—Ç–æ –∫—Ä–∏–≤—ã–µ —Ç–µ–ø–µ—Ä—å –Ω–∞—á–∏–Ω–∞—é—Ç –¥–≤–∏–∂–µ–Ω–∏–µ —Å —Ç–æ—á–∫–∏ $(1,1)$ –∏ –∑–∞–∫–∞–Ω—á–∏–≤–∞—é—Ç –≤ —Ç–æ—á–∫–µ $(0,0)$. –ù–æ —Å—Ç–æ–∏—Ç —É—Ç–æ—á–Ω–∏—Ç—å, —á—Ç–æ –∫—Ä–∏–≤—ã–µ —Ç–æ–ª—å–∫–æ —Å—Ç—Ä–µ–º—è—Ç—Å—è –∫ –¥–∞–Ω–Ω—ã–º –≥—Ä–∞–Ω–∏—á–Ω—ã–º —Ç–æ—á–∫–∞–º –∏ –¥–æ—Å—Ç–∏–≥–Ω—É—Ç –∏—Ö —Ç–æ–ª—å–∫–æ –ø—Ä–∏ $Œª=-‚àû$ –∏ –ø—Ä–∏ $Œª=‚àû$. –≠—Ç–æ —Å–≤—è–∑–∞–Ω–æ —Å —Ç–µ–º, —á—Ç–æ —É—Å–ª–æ–≤–Ω—ã–µ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ $(6)$ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–∞ $\mathbb{R}$ –∏ –∏–Ω—Ç–µ–≥—Ä–∞–ª –æ—Ç –Ω–∏—Ö —Ä–∞–≤–µ–Ω $1$ –∏–ª–∏ $0$ —Ç–æ–≥–¥–∞ –∏ —Ç–æ–ª—å–∫–æ —Ç–æ–≥–¥–∞, –∫–æ–≥–¥–∞ $Œª=-‚àû$ –∏–ª–∏ $Œª=‚àû$ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ.

–í—Ç–æ—Ä–æ–µ –æ–±—ä—è—Å–Ω—è–µ—Ç—Å—è —Ç–µ–º, —á—Ç–æ –º–æ–∂–Ω–æ –¥–æ—Å—Ç–∏—á—å –±–æ–ª—å—à–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è $P_d$ –ø—Ä–∏ –º–µ–Ω—å—à–µ–º –∑–Ω–∞—á–µ–Ω–∏–∏ $P_{fa}$. –≠—Ç–æ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –ø–æ—Ç–æ–º—É, —á—Ç–æ –ø—Ä–∏ —É–≤–µ–ª–∏—á–µ–Ω–∏–∏ –∑–Ω–∞—á–µ–Ω–∏—è $SNR$, —É–º–µ–Ω—å—à–∞–µ—Ç—Å—è –∑–Ω–∞—á–µ–Ω–∏–µ $œÉ$, —á—Ç–æ –≤–∏–¥–Ω–æ –∏–∑ $(9)$. –ü–æ—Å–º–æ—Ç—Ä–∏–º, –∫–∞–∫ –≤–∏–∑—É–∞–ª—å–Ω–æ –∏–∑–º–µ–Ω—è—é—Ç—Å—è –∫—Ä–∏–≤—ã–µ —É—Å–ª–æ–≤–Ω—ã—Ö –ø–ª–æ—Ç–Ω–æ—Å—Ç–µ–π $(6)$ –ø—Ä–∏ —Ä–∞–∑–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏—è—Ö $SNR$.

```python
# —Ñ—É–Ω–∫—Ü–∏—è, –∫–æ—Ç–æ—Ä–∞—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ —É—Å–ª–æ–≤–Ω—ã—Ö –ø–ª–æ—Ç–Ω–æ—Å—Ç–µ–π –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç ùùà
def conditional_density(x, sigma):
    return norm.pdf(x/sigma), norm.pdf((x - 1)/sigma)
```

```python
# —Å–æ–∑–¥–∞–Ω–∏–µ DataFrame –≤ –∫–æ—Ç–æ—Ä–æ–º –±—É–¥—É—Ç —Ö—Ä–∞–Ω–∏—Ç—å—Å—è –∑–Ω–∞—á–µ–Ω–∏—è —É—Å–ª–æ–≤–Ω—ã—Ö –ø–ª–æ—Ç–Ω–æ—Å—Ç–µ–π –∏ SNR
data3 = []
x_range = np.linspace(-5, 5, 100)
for SNR in SNR_values:
    for x in x_range :
        cond_dens1, cond_dens2 = conditional_density(x, snr_to_sigma(SNR))
        data3.append([cond_dens1, cond_dens2, str(SNR), x])
        
data3 = pd.DataFrame(data3, columns=['cond_dens1', 'cond_dens2', 'SNR', 'x'])
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
sns.set_context('poster')
plt.title('–£—Å–ª–æ–≤–Ω—ã–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –∑–Ω–∞—á–µ–Ω–∏—è $SNR$', 
          fontsize=30)
plt.xlabel('$x$', fontsize=40)
plt.xticks(np.arange(-5, 5.5, 0.5), fontsize=20)
plt.yticks(np.arange(0, 0.45, 0.05), fontsize=20)
plt.ylabel('–ó–Ω–∞—á–µ–Ω–∏–µ —É—Å–ª–æ–≤–Ω–æ–π –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏', fontsize=20, 
           labelpad=30)

sns.lineplot(x='x', y='cond_dens2', hue='SNR', 
             data=data3, palette='gist_ncar', ax=ax)

sns.lineplot(x='x', y='cond_dens1', hue='SNR', 
             data=data3, palette='gist_ncar', ax=ax, legend=False)

ax.legend(fontsize='small', shadow=True, title='$SNR(dB)$',
           borderpad=0.9, fancybox=True, loc='upper left')

ax.annotate("$f_{X|H_0}(x|H_0)$", xy=(-0.6, 0.35),  xycoords='data',
            xytext=(-90, 20), textcoords='offset points',
            size=30, ha='right', 
            arrowprops=dict(arrowstyle="->", color='black',
                            connectionstyle="arc3,rad=0.1"))

ax.annotate("$f_{X|H_1}(x|H_1)$", xy=(1.6, 0.35),  xycoords='data',
            xytext=(100, 20), textcoords='offset points',
            size=30, ha='left', 
            arrowprops=dict(arrowstyle="->", color='black',
                            connectionstyle="arc3,rad=0.1"))

plt.grid(color='black', linestyle='--', linewidth=1, alpha=0.3);
```

–ü–æ –†–∏—Å. 7 –≤–∏–¥–Ω–æ, —á—Ç–æ –ø—Ä–∏ —É–≤–µ–ª–∏—á–µ–Ω–∏–∏ $SNR$, –∫—Ä–∏–≤—ã–µ —Å—Ç–∞–Ω–æ–≤—è—Ç—Å—è —É–∂–µ, –∞, –∑–Ω–∞—á–∏—Ç, –ø–µ—Ä–µ—Å–µ–∫–∞—é—Ç—Å—è –≤ –º–µ–Ω—å—à–µ–π —Å—Ç–µ–ø–µ–Ω–∏. –ü–æ—Å–º–æ—Ç—Ä–∏–º —Ç–µ–ø–µ—Ä—å –Ω–∞ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏—è $P_d$ –æ—Ç $Œª$:

```python
data4 = []
for SNR in SNR_values:
    for lambda_ in lambda_values:
        Pfa, Pd = roc_curve(snr_to_sigma(SNR), lambda_)
        data4.append([Pfa, Pd, str(SNR), lambda_])
        
data4 = pd.DataFrame(data4, columns=['Pfa', 'Pd', 'SNR', 'lambda'])
```

–ü—Ä–∏ $Œª=0$ –ø–æ–ª—É—á–∞–µ—Ç—Å—è, —á—Ç–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ª–æ–∂–Ω–æ–π —Ç—Ä–µ–≤–æ–≥–∏ —Ä–∞–≤–Ω–∞ $\frac{1}{2}$, —Ç–∞–∫ –∫–∞–∫ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã–π –∏–Ω—Ç–µ–≥—Ä–∞–ª –æ—Ç —É—Å–ª–æ–≤–Ω–æ–π –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è $
\left.\frac{1}{\sqrt{2 \pi \sigma^{2}}} \int_{\lambda}^{+\infty} e^{-\frac{x^{2}}{2 \sigma^{2}}} d x\right|_{\lambda=\eta_{0}}
$ —Å—á–∏—Ç–∞–µ—Ç—Å—è –æ—Ç –∑–Ω–∞—á–µ–Ω–∏—è –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è –¥–æ $+‚àû$, —á—Ç–æ –≤ —Ç–æ—á–Ω–æ—Å—Ç–∏ —Ä–∞–≤–Ω–æ $\frac{1}{2}$, –≤ —Å–∏–ª—É —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ—Å—Ç–∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.

–¢—Ä–µ—Ç—å–µ –æ–±—ä—è—Å–Ω—è–µ—Ç—Å—è —Ç–µ–º, —á—Ç–æ —Ç–∞–Ω–≥–µ–Ω—Å —É–≥–ª–∞ –Ω–∞–∫–ª–æ–Ω–∞ $ROC$-–∫—Ä–∏–≤–æ–π –≤ –Ω–µ–∫–æ—Ç–æ—Ä–æ–π –µ–µ —Ç–æ—á–∫–µ —Ä–∞–≤–µ–Ω –∑–Ω–∞—á–µ–Ω–∏—é –ø–ª–æ—Ç–Ω–æ—Å—Ç–µ–π $(6)$ –æ—Ç –ø–æ—Ä–æ–≥–∞ –¥–µ—Ç–µ–∫—Ç–∏—Ä–æ–≤–∞–Ω–∏—è $Œª$, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ–º—É –¥–ª—è –¥–æ—Å—Ç–∏–∂–µ–Ω–∏—è $P_d$ –∏ $P_{fa}$  –≤ —ç—Ç–æ–π —Ç–æ—á–∫–µ.

–ü—É—Å—Ç—å

$$
\tan \left(\alpha_{\lambda}\right):=\frac{\left(\frac{d P_{d}}{d \lambda}\right)}{\left(\frac{d P_{f a}}{d \lambda}\right)} \ \ ,(14)
$$

—Ç–æ–≥–¥–∞:

$$
\begin{aligned}
&\tan \left(\alpha_{\lambda}\right)=\frac{\frac{d}{d \lambda}\left(\frac{1}{\sqrt{2 \pi \sigma^{2}}} \int_{\lambda}^{+\infty} e^{-\frac{(x-1)^{2}}{2 \sigma^{2}}} d x\right)}{\frac{d}{d \lambda}\left(\frac{1}{\sqrt{2 \pi \sigma^{2}}} \int_{\lambda}^{+\infty} e^{-\frac{x^{2}}{2 \sigma^{2}}} d x\right)}=\frac{-\frac{1}{\sqrt{2 \pi \sigma^{2}}} e^{-\frac{(\lambda-1)^{2}}{2 \sigma^{2}}}}{-\frac{1}{\sqrt{2 \pi \sigma^{2}}} e^{-\frac{(\lambda)^{2}}{2 \sigma^{2}}}}=\\
&=\frac{e^{-\frac{(\lambda-1)^{2}}{2 \sigma^{2}}}}{e^{-\frac{(\lambda)^{2}}{2 \sigma^{2}}}}=e^{-\frac{(\lambda-1)^{2}}{2 \sigma^{2}}+\frac{(\lambda)^{2}}{2 \sigma^{2}}}=e^{\frac{-\lambda^{2}+2 \lambda-1+\lambda^{2}}{2 \sigma^{2}}}=e^{\frac{2 \lambda-1}{2 \sigma^{2}}} \ \ \ (15)
\end{aligned} 
$$

–ò—Å—Å–ª–µ–¥—É–µ–º —Ñ—É–Ω–∫—Ü–∏—é $(15)$ :

–ò–∑ $(15)$  –∏ $(16)$ —Å–ª–µ–¥—É–µ—Ç, —á—Ç–æ —Ç–∞–Ω–≥–µ–Ω—Å —É–≥–ª–∞ –Ω–∞–∫–ª–æ–Ω–∞ $ROC$-–∫—Ä–∏–≤–æ–π –º–æ–Ω–æ—Ç–æ–Ω–Ω–æ –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –æ—Ç $0$ –¥–æ $+‚àû$ –ø—Ä–∏ $Œª$ –æ—Ç $-‚àû$ –¥–æ $+‚àû$. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –Ω–∞—á–∏–Ω–∞—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ –æ—Ç $Œª=-‚àû \ $  $tan‚Å°(Œ±_Œª )=0$ –∏, —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ, $ Œ±_Œª=0¬∞$. –ó–∞–∫–∞–Ω—á–∏–≤–∞–µ—Ç—Å—è –≤—Å–µ –≤ —Ç–æ—á–∫–µ $Œª=+‚àû$, –∫–æ–≥–¥–∞ $tan‚Å°(Œ±_Œª )=+‚àû$, —Ç.–µ. $Œ±_Œª=90¬∞$ (–∫–∞—Å–∞—Ç–µ–ª—å–Ω–∞—è –≤ –¥–∞–Ω–Ω–æ–π —Ç–æ—á–∫–µ –ø–µ—Ä–ø–µ–Ω–¥–∏–∫—É–ª—è—Ä–Ω–∞ –æ—Å–∏ –∞–±—Å—Ü–∏—Å—Å)


```python
# —Ñ—É–Ω–∫—Ü–∏—è, –∫–æ—Ç–æ—Ä–∞—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏—è Pfa –∏ Pd –ø–æ –∑–∞–¥–∞–Ω–Ω—ã–º ùùà –∏ Œª
def dif_roc_curve(sigma, lambda_):
    return (derivative(lambda x: 1 - norm.cdf(x/sigma), lambda_, dx=1e-10),
           derivative(lambda x: 1 - norm.cdf((x - 1)/sigma), lambda_, dx=1e-10))


data5 = []
for SNR in SNR_values:
    for lambda_ in [0, 0.5, 1]:
        dPfa, dPd = dif_roc_curve(snr_to_sigma(SNR), lambda_)
        Pfa, Pd = roc_curve(snr_to_sigma(SNR), lambda_)
        tan = dPd/dPfa
        for i in np.arange(0.01/tan, 0.2/tan, 0.01/tan):
            data5.append([tan*(i - Pfa) + Pd, i, str(SNR), lambda_])
        
data5 = pd.DataFrame(data5, columns=['y', 'x', 'SNR', 'lambda'])
```
